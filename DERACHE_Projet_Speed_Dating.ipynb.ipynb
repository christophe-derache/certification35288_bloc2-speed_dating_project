```python
def check_create_lists(df):
    list_latitude = []
    list_longitude= []
    
    
    for area in df['origin_location']:
        if area in dict_lat_lon:
            list_latitude.append(float(dict_lat_lon[area]['latitude']))
            list_longitude.append(float(dict_lat_lon[area]['longitude']))
            
        else:
            list_latitude.append(float('0'))
            list_longitude.append(float('0'))
    
    return list_latitude, list_longitude
                
```

```python
list_latitude, list_longitude = check_create_lists(df_from_counts)

df_from_counts['latitude'] = list_latitude
df_from_counts['longitude'] = list_longitude
```

### 2 - DATA CLEANING

#### a) Deletion of rows when 'Wave' is between 6 and 9

```python
df = df.drop(df[(df.wave < 10) & (df.wave > 5)].index)
```

```python
df['wave'].unique()
```

#### b) Deletion of the columns without interest for the study or containing too few data

##### Defining column list

```python
def create_a_list_to_drop(df, value):
    list_to_drop = []
    
    for col in df.columns:
        
        sum_null = df[col].isnull().sum()
        if sum_null >= value:
            list_to_drop.append(col)
        
    return list_to_drop
        
```

```python
columns_containing_too_few_data = create_a_list_to_drop(df, 4000)
len(columns_containing_too_few_data)
```

```python
columns_without_interest = ['id', 'idg', 'condtn', 'wave', 'round', 'position', 'positin1', 'order', 
                            #'samerace', 
                            'race_o', 'field', 'field_cd', 'undergra', 'mn_sat', 'tuition', 'race', 
                            'imprace', 'from', 'zipcode', 'income']
len(columns_without_interest)
```

```python
columns_to_drop = columns_containing_too_few_data + columns_without_interest
len(columns_to_drop)
```

##### Dropping columns

```python
df.drop(columns_to_drop, axis=1, inplace=True)
```
